```python
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV, KFold
```

```python
folds = KFold(n_splits=10, shuffle=True)
```

```python
elasticModel = ElasticNet()
```

```python
params = {
    'l1_ratio': [0.0, 0.3, 0.5, 0.8, 1.0],
    'alpha': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]
}
search = GridSearchCV(elasticModel, param_grid=params, cv=folds, scoring='r2')
search.fit(x_train, y_train)
```

```python
display(search.best_score_)
display(search.best_params_)
```

```python
yhat = search.predict(x_test)
dt.rSquare(x_test, y_test, yhat)
```

```python
dt.coef(search.best_estimator_, x_train)
```

```python
resid = y_test-yhat
sns.histplot(resid)
```
