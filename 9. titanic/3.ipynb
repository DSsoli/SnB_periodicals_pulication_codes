```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold, cross_val_score
```

```python
x = rawData.drop(columns=['Survived'])
y = rawData['Survived']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True, random_state=123)
```

```python
folds = StratifiedKFold(n_splits=10, shuffle=True, random_state=123)
```

```python
baseModel = LogisticRegression(max_iter=100000000)
baseModel.fit(x_train, y_train)
```

```python
yhat_train = baseModel.predict(x_train)
yhat_test = baseModel.predict(x_test)
```

```python
print(classification_report(y_train, yhat_train))
```

```python
print(classification_report(y_test, yhat_test))
```

```python
scores_train = cross_val_score(baseModel, x_train, y_train, scoring='roc_auc', cv=folds)
scores_test = cross_val_score(baseModel, x_test, y_test, scoring='roc_auc', cv=folds)
```

```python
display(scores_train.mean())
display(scores_test.mean())
```

```python
import eli5
eli5.show_weights(baseModel, feature_names = x_train.columns.tolist())
```

```python
from sklearn.svm import SVC
```

```python
svmModel = SVC()
params = {
    'C': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000],
    'gamma': [0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000]
}
search = GridSearchCV(svmModel, param_grid=params, cv=folds, scoring='roc_auc')
search.fit(x_train, y_train)
```

```python
display(search.best_params_)
display(search.best_estimator_)
display(search.best_score_)
```

```python
yhat_train = search.predict(x_train)
yhat_test = search.predict(x_test)
```
