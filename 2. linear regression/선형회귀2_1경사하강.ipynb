```python
def error(x):
  return x ** 2 - 5 * x + 10

def grad_fx(x):
  return 2 * x - 5
```

```python
x0 = -4
lr = 0.01
maxIter = 1000
for i in range(maxIter):
  x1 = x0 - lr * grad_fx(x0)
  x0 = x1
print('error: {}, grad: {}, x: {}'.format(error(x0), grad_fx(x0), x0))
```

```python
def error(a, b):
  return (a - 2) ** 2 + (b - 2) ** 2

def grad_fa(a, b):
  return 2 * a + 8

def grad_fb(a, b):
  return 2 * b + 6
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(projection='3d')

xx1, xx2 = np.meshgrid(np.arange(-80, 80), np.arange(-80, 80))
ER = error(xx1, xx2)
ax.plot_surface(xx1, xx2, ER, color='g')
plt.show()
```

```python
beta0 = np.array([-80, 80])
lr = 0.001
maxIter = 1000
DC = []
for _ in range(maxIter):
  DC.append(beta0)
  grad = np.array([ grad_fa(*beta0), grad_fb(*beta0) ])
  beta1 = beta0 - lr * grad
  beta0 = beta1
```

```python
fig = plt.figure(figsize = (10, 10))
ax = fig.add_subplot(projection='3d')

xx1, xx2 = np.meshgrid(np.arange(-80, 80), np.arange(-80, 80))
ER = error(xx1, xx2)
ax.plot_surface(xx1, xx2, ER, alpha=0.4, color='r')
ax.plot(np.array(DC)[:,0], np.array(DC)[:,1], error(np.array(DC)[:,0], np.array(DC)[:,1]), 'o-', color='b')
plt.show()
```
