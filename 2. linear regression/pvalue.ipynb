```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
stdRv = norm(loc=0, scale=1)
x = np.linspace(-5, 5, 1000)
x2 = np.linspace(-9.35, 0.65, 1000)
plt.plot(x, stdRv.pdf(x))
plt.axvline(x=4.7, ymax=0.004, color='r', linewidth=20)
plt.axvline(x=4.35, ymax=0.2, color='g', linewidth=2, linestyle=':')
plt.axvline(x=-4.7, ymax=0.004, color='r', linewidth=20)
plt.axvline(x=-4.35, ymax=0.2, color='g', linewidth=2, linestyle=':')
plt.plot(x2, stdRv.pdf(x))
plt.yticks([])
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
stdRv = norm(loc=0, scale=1)
plt.plot(x, stdRv.pdf(x))
plt.axvline(x=4.7, ymax=0.004, color='r', linewidth=20)
section = np.arange(:-4.35)
plt.fill_between(section, stdRv.pdf(section))
plt.axvline(x=4.35, ymax=0.2, color='g', linewidth=2, linestyle=':')
plt.yticks([])
plt.grid()
plt.show()
```
