```python
from scipy.stats import bernoulli
seed = 5000
p = 0.45
balance = []
bet = [-150, 150]

while True:
    i = bernoulli(p).rvs(1)[0]
    seed += bet[i]
    balance.append(seed)
    if seed<=0: 
        break
    
```

```python
import matplotlib.pyplot as plt
plt.plot(balance)
plt.grid()
plt.show()
```

```python
ngames = []
p = 0.45
bet = [-150, 150]

for _ in range(500):
    seed = 5000
    cnt = 0
    while True:
        cnt += 1
        i = bernoulli(p).rvs(1)[0]
        seed += bet[i]
        if seed<=0:
            break
    ngames.append(cnt)
```

```python
import numpy as np
display(np.mean(ngames))
display(np.var(ngames))
```

```python
import seaborn as sns
sns.distplot(ngames)
plt.grid()
plt.show()
```
