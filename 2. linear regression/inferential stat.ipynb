```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
x = np.array([60, 65, 69, 75, 82, 88])
y = np.array([165, 168, 174, 177, 179, 183])
x = x.reshape(-1,1)
```

```python
plt.plot(x, y, 'o')
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid()
plt.show()
```

```python
linearModel = LinearRegression().fit(x, y)
display(linearModel.coef_)
display(linearModel.intercept_)
```

```python
fitline=linearModel.coef_*x+linearModel.intercept_
```

```python
meanline = np.mean(y)+0*x
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y,'o')
plt.plot(x, fitline, label='Linear Model')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x, y, 'o')
plt.plot(x, meanline, label='Mean of Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```

```python
y2 = linearModel.predict(x)
meanline2 = np.mean(y2) + 0*x
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.plot(x,y2,'o')
plt.plot(x, fitline, label='Linear Model')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.subplot(1,2,2)
plt.plot(x, y2, 'o')
plt.plot(x, meanline2, label='Mean of Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()

plt.show()
```

```python
linearModel2 = LinearRegression().fit(x, y2)
linearModel2.score(x, y2)
```
