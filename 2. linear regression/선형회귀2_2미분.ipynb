```python
import numpy as np
import sympy as sp

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def error(x):
    return x**2 - 5*x + 10
```

```python
x = sp.Symbol('x')
f = x ** 2 - 5 * x + 10
sp.diff(f, x)
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, error(x))
plt.grid()
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, error(x))
plt.plot(2.5, error(2.5), 'o')
plt.grid()
plt.show()
```

```python
x = np.linspace(-5, 5, 100)
plt.plot(x, error(x))
plt.plot(2.5, error(2.5), 'o', label='Least error: "differential coefficient=0"')
plt.plot(x, 0*x+error(2.5), label='instantaneous slope=0')
plt.legend()
plt.grid()
plt.show()
```
